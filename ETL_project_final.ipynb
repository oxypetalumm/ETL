```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
sbdrinks_csv = "Resources/starbucks_drinks.csv"
mcdfood_csv = "Resources/mcd_menu.csv"
swfood_csv = "Resources/sw_food.csv"


starbucks_drinks_df = pd.read_csv(sbdrinks_csv, na_values=['-'])
mcd_df = pd.read_csv(mcdfood_csv)
sw_menu_df = pd.read_csv(swfood_csv)
```

```python
display(starbucks_drinks_df.head())
display(mcd_df.head())
display(sw_menu_df.head())
```

```python
starbucks_drinks_df.columns
```

```python
new_starbucks_drinks_df = starbucks_drinks_df.rename(columns ={"Unnamed: 0": "item", 'Calories': 'calories', 'Fat (g)' : 'total_fat', 'Carb. (g)': 'carbohydrates', 'Protein': 'protein'})
```

```python
new_starbucks_drinks_df.columns
```

```python
new_starbucks_drinks_df.head()
```

```python
new_starbucks_drinks_df = new_starbucks_drinks_df[['item', 'calories', 'total_fat', 'carbohydrates', 'protein']]
new_starbucks_drinks_df.head()
```

```python
new_starbucks_drinks_df.isnull().sum()
```

```python
new_starbucks_drinks_df.dtypes
```

```python
new_mcd_df = mcd_df[['Item', 'Calories', 'Total Fat', 'Carbohydrates', 'Protein']]
new_mcd_df = new_mcd_df.rename(columns ={"Item":"item", "Total Fat": "total_fat", 'Calories': 'calories', "Carbohydrates": "carbohydrates", "Protein": "protein" })
new_mcd_df.head()
```

```python
sw_menu_df.columns
```

```python
new_sw_menu_df = sw_menu_df.rename(columns ={"Unnamed: 0": "item", 'Calories': 'calories', "Total Fat (g)": "total_fat", "Carbohydrates (g)": "carbohydrates", "Protein (g)": "protein"})
```

```python
new_sw_menu_df.columns
```

```python
new_sw_menu_df = new_sw_menu_df[['item', 'calories', 'total_fat', 'carbohydrates', 'protein']]
new_sw_menu_df.head()
```

```python
rds_connection_string = "postgres:<password>@localhost:5432/starbucks_menu"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_starbucks_drinks_df.to_sql(name='starbucks_menu_drinks', con=engine, if_exists='replace', index=False)
```

```python
new_mcd_df.to_sql(name='mcd_menu', con=engine, if_exists='replace', index=False)
```

```python
new_sw_menu_df.to_sql(name='sw_menu', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from starbucks_menu_drinks', con=engine).head()
```

```python
pd.read_sql_query('select * from mcd_menu', con=engine).head()
```
